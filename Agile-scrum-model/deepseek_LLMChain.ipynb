```python
import os
import nest_asyncio
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain_community.llms import Replicate
from constants import Constants
nest_asyncio.apply()

# Set API token
os.environ["REPLICATE_API_TOKEN"] = Constants.REPLICATE_API_KEY 
```

```python
# Initialize compatible LLM for LangChain
llm = Replicate(
    model="deepseek-ai/deepseek-r1",
    model_kwargs={"temperature": 0.7, "top_p": 1},
    replicate_api_token=os.environ["REPLICATE_API_TOKEN"]
)

# Role-based prompts
sprint_planning_prompt = PromptTemplate.from_template(
    "You are a Product Owner. Prepare the sprint backlog with user stories, priorities, and estimate the time (in days) needed for each user story. Also, provide the resources required, expected completion date, and the estimated number of sprints to complete the entire project."
)

scrum_master_prompt = PromptTemplate.from_template(
    "You are a Scrum Master. Based on the backlog, plan the sprint execution steps. Estimate the time (in days) for each task, provide necessary resources, expected finish date, and estimate the number of sprints required to complete the project."
)

developer_prompt = PromptTemplate.from_template(
    "You are a Developer. Implement the user stories as per the sprint plan. Estimate how much time you will need for each task and what resources (tools, team members) you would require. Also, estimate how many sprints you'll need to complete your tasks."
)

qa_prompt = PromptTemplate.from_template(
    "You are a QA. Test the developed user stories. Estimate how long testing will take, what resources (team members, tools) you need, and suggest the expected finish date. How many sprints will you require to complete the testing?"
)

review_prompt = PromptTemplate.from_template(
    "You are a Review Board. Evaluate the sprint execution and provide feedback. Estimate how long the review will take, required resources, and expected finish date. Estimate how many sprints will be required to complete the project review."
)

business_analyst_prompt = PromptTemplate.from_template(
    "You are a Business Analyst. Based on the sprint backlog, identify and refine business requirements, align user stories with business goals, and estimate the number of sprints needed to finalize all requirements."
)

designer_prompt = PromptTemplate.from_template(
    "You are a UI/UX Designer. Design user interfaces and user experiences for each user story. Estimate how long the design will take, required resources, tools, and how many sprints are needed."
)

devops_prompt = PromptTemplate.from_template(
    "You are a DevOps Engineer. Plan CI/CD pipelines, deployment environments, and infrastructure setup. Estimate time, resources, and number of sprints required."
)

architect_prompt = PromptTemplate.from_template(
    "You are a System Architect. Define the system architecture based on the sprint backlog. Estimate design duration, resources, and number of sprints required to complete architecture planning."
)

security_engineer_prompt = PromptTemplate.from_template(
    "You are a Security Engineer. Identify security risks and mitigation strategies for user stories. Estimate the time and resources required, and number of sprints needed."
)

technical_writer_prompt = PromptTemplate.from_template(
    "You are a Technical Writer. Prepare documentation (user guides, developer docs) for each completed user story. Estimate time, tools, and number of sprints needed."
)


# Initialize chains for each role
sprint_chain = LLMChain(llm=llm, prompt=sprint_planning_prompt)
scrum_master_chain = LLMChain(llm=llm, prompt=scrum_master_prompt)
developer_chain = LLMChain(llm=llm, prompt=developer_prompt)
qa_chain = LLMChain(llm=llm, prompt=qa_prompt)
review_chain = LLMChain(llm=llm, prompt=review_prompt)
business_analyst_chain = LLMChain(llm=llm, prompt=business_analyst_prompt)
designer_chain = LLMChain(llm=llm, prompt=designer_prompt)
devops_chain = LLMChain(llm=llm, prompt=devops_prompt)
architect_chain = LLMChain(llm=llm, prompt=architect_prompt)
security_chain = LLMChain(llm=llm, prompt=security_engineer_prompt)
writer_chain = LLMChain(llm=llm, prompt=technical_writer_prompt)

# Run the simulation by invoking each chain with detailed output
print("========== Agile Sprint Simulation ==========\n")

response = sprint_chain.invoke({})
print("Sprint Planning:\n", response['text'].strip(), "\n")

print("Business Analyst:\n", business_analyst_chain.invoke({})['text'].strip(), "\n")
print("UI/UX Designer:\n", designer_chain.invoke({})['text'].strip(), "\n")
print("Architect:\n", architect_chain.invoke({})['text'].strip(), "\n")
print("Scrum Master:\n", scrum_master_chain.invoke({})['text'].strip(), "\n")
print("Developer:\n", developer_chain.invoke({})['text'].strip(), "\n")
print("Security Engineer:\n", security_chain.invoke({})['text'].strip(), "\n")
print("DevOps:\n", devops_chain.invoke({})['text'].strip(), "\n")
print("QA:\n", qa_chain.invoke({})['text'].strip(), "\n")
print("Technical Writer:\n", writer_chain.invoke({})['text'].strip(), "\n")
print("Sprint Review:\n", review_chain.invoke({})['text'].strip(), "\n")

```
